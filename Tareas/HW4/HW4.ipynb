```python
#Importar la documentación necesaria 
%pylab inline 
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
from scipy.fftpack import ifft, fft, fftfreq
```

```python
#Crear mi arreglo vacio con ceros 
#Con la función que pertence a numpy (np.zeros) puedio crear un array lleno de ceros pero en una sola dimensión, entonces con la opción de np.resize creo la matriz
arregloCeros=np.zeros(1024**2)
arregloCeros=np.resize(arregloCeros,(1024,1024))
print arregloCeros
```

```python
#En el primer for recorro las posiciones en x
#En el segundo for recorro las posiciones en y
#Para el primer if le estoy dando la posición en todo el centro de mi array inciial que es 512(Centro del circulo), este debe ser de 32x32 lo que significa que en x debe estar entre 496 y 528
#Creo cuadrado para saber que ahora este es el array que corresponde al que tiene el cuadrado en centro lleno de números 2
#NO oolvidar que debo restar -1 a cada posicion debido a que los arrays comienzana en cero
cuadrado=arregloCeros
for i in range(496,528):
    for j in range(496,528):
            cuadrado[i][j]=1
print cuadrado

#Con la función np.count_nonzero me cuenta los no ceros que se encuentran dentrode mi matriz 
cuenta=np.count_nonzero(cuadrado)
print cuenta

#Mostrar mi arreglo, de esta forma veo mi cuadrado
imshow(cuadrado)
```

```python
#Aplicar la transformada discreta de Fourier 
cuadradTrans=np.fft.fft2(cuadrado)
print cuadradTrans
```

```python
coord=sp.fftpack.fftshift(cuadradTrans)
print coord
```

```python
#Requiero el absoluto del arreglo que ya tiene cambiada las coordenadas 
absoluto=abs(coord)
imshow(absoluto)
```

```python
#Sacar cuadrado de 200 de lado cone el for 
#Se aplica el módulo de la función sacando la raiz cuadrada a la matriz eleveda al cuadrado
modulo=(sqrt((absoluto)**2))
logMod=log(modulo)

final=np.zeros((200,200))
for i in range(412,612):
    for j in range(412,612):
        final[i-412][j-412]=logMod[i][j]

plt.title("Patron de Interferencia de una apertura cuadrada")
imshow(final)
```

```python
#Para una abertura de radio 0.01
#Crear mi arreglo vacio con ceros 
arregloCeros3=np.zeros((1024,1024))
print arregloCeros3
```

```python
#Para crear el circulo
#Si hago la equivalencia de 1024 como una unidad a 0,01 son aproximadamente 10 pixeles 

circle=arregloCeros3

for i in range(502,522):
    for j in range(502,522):
        if(((i-512)**2) + ((j-512)**2)<=104.8):
            circle[i][j]=1
print circle      
imshow(circle)
```

```python
#Aplicar la transformada discreta de Fourier a los dos arreglos
circuloTrans=np.fft.fft2(circle)
print circuloTrans
```

```python
#Función para el cambio de coordenadas
coord2=sp.fftpack.fftshift(circuloTrans)
print coord2
```

```python
#Requiero el absoluto de los arreglos que ya tiene cambiada las coordenadas 
absoluto3=(abs(coord2))
absoluto3=(absoluto3**2)
imshow(np.log(absoluto3+1))
plt.figure(figsize=(20,15))
```

```python
#Para poder graficar la Intesidad Vs X.
#Creo un variable de las filas centrales, luego otra de la variable con un maximo
#Un array con las x de las abcsisas con la ayuda de np.linspace 
absol= abs(coord2)
central=absol[512]
central=(central)**2
maxiN=(coord2[512][512])**2
InNorm=central/maxiN
x=2*np.pi*(0.01)/1.*np.linspace(-512,512,1024)
plt.scatter(x, InNorm)
plt.xlim(0,10)
plt.ylim(0,1.0)
```

```python
#Con las recomendaciones dadas en la guia, es posible crear la cuadrícula 
#Después de cada comando de creación de una grid re-defino la linea de código de cada gráfica y así hago con las tres gráficas
grafica=plt.figure(figsize=(15, 10))
plt.subplot2grid((3,2),(0,0), colspan=2, rowspan=1)
plt.scatter(x, InNorm)
plt.xlim(0,10)
plt.xlabel("$2 \pi Rj/ L$")
plt.ylabel("$I/I_0$")
plt.subplot2grid((3,2),(1,0), rowspan=2)
imshow(circle)
plt.subplot2grid((3,2),(1,1),rowspan=2)
imshow(np.log(absoluto3+1))
plt.title("Patron de interferencia de una apertura circular (Patron de Airy)\n log resaltado")
grafica.tight_layout()
plt.show()
```

```python
#Hallar máximo y mínimo de la gráfica
#Ahora tiene un nuevo rango la gráfica de Intensidad Vs X, va de 0 a 12 en posición 
#Para este nuevo rango filtro en la primera variable los positivos y luego filtro hasta 12
plt.scatter(x, InNorm)
plt.title("Intensidad Vs X - Completa")
plt.xlim(0,10)
positivos=x[x>0]
absicisaLim=positivos[positivos<12]
```

```python
#Inclusivo con el primero, exlcusivo con el último, regla fundamental de python!!
#Con la ayuda de np.where puedo conocer la posición donde me encuentro y puedo definir mis nuevos limites que van a estar incluidos en la variable nuevaInNorm

beginn=np.where(x == absicisaLim[0])[0]

finale=np.where(x == absicisaLim[len(absicisaLim)-1])[0]

nuevaInNomr=InNorm[beginn:finale+1]
```

```python
#Con la ayuda de la herramienta slice notation
#ReaLizo esto como un arreglo de Booleam, como si fueran tablas de verdad y pongo la condición de maximo 
#Luego hallo la posiicion y el valor en esa posición de los puntos máximos
maximos=np.r_[True, nuevaInNomr[1:] > nuevaInNomr[:-1]] & np.r_[nuevaInNomr[:-1] > nuevaInNomr[1:], True]

maxPosic = np.where(maximos ==True)[0]

valorMax = nuevaInNomr[maxPosic]

print maxPosic
print valorMax
```

```python
#Para hallar mínimos es un proceso 
#Con una tabal de valor comparo el indice y hallo los minimos
#Con la herramienta np.where hallo en donde se encuntran mis valores y lugo sus valores reales
minimos=np.r_[True, nuevaInNomr[1:] < nuevaInNomr[:-1]] & np.r_[nuevaInNomr[:-1] < nuevaInNomr[1:], True]

minPosic = np.where(minimos == True)[0]

valorMin = nuevaInNomr[minPosic]

print minPosic
print valorMin
```

```python
#Interpolación de Grado 10, fue la interpolación que más se asemeja a mi gráfica 
fit=np.polyfit(absicisaLim, nuevaInNomr, 10)
interpol= np.poly1d(fit)

x=2*np.pi*(0.01)/1.*np.linspace(-512,512,1024)
newX=np.linspace(0,12,1000)

plt.plot(newX,interpol(newX), color='yellow')
plt.xlim(0,10)
plt.scatter(x, InNorm , color='k')
plt.title('Intensidad Vs X con Interpolacion')
plt.xlabel("$2 \pi Rj/ L$")
plt.ylabel("$I/I_0$")
```
