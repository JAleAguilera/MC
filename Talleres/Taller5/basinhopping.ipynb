```python
#Para la constante, debo crear la función en cuestión que esta descrita como el seno cardinal
def sinCardi(t,a):
    return (sin((2*a+1)*t)/(2*a+1))
```

```python
#La función para el ajuste, es la misma que la del seno cardinal pero esta es acumulativa.
def ajuste(t,num):
    final = 0
    for i in range(num):
        final=final+sinCardi(t,i)
    return final
```
